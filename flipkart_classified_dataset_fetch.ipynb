```python
#This code provides functionality of a webscraper to fetch data from flipkart classified in various labels.

import pandas as pd
import requests
import os
import matplotlib.pyplot as plt
import io
from PIL import Image
from pathlib import Path
# hashlib allows you to get hashes. Let's use 'sha1' to name the images.
import hashlib


from bs4 import BeautifulSoup


import csv
```

```python
url = "https://www.flipkart.com/search?q=shoes&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
import requests
from bs4 import BeautifulSoup

req = requests.get(url)
htmldata=req.text
soup = BeautifulSoup(htmldata, 'html.parser')
# <a class="ge-49M" href="/clothing-and-accessories/pr?sid=clo&amp;otracker=categorytree&amp;page=1">1</a>

results=[]

for a in soup.findAll(attrs={"class": "_312yBx SFzpgZ"}):
    name = a.find("img")
    if name not in results:
        results.append(name.get("src"))
                           
                           
```

```python
i =0
for b in results:
        if(i==15):
            break
        image_content = requests.get(b).content
        image_file = io.BytesIO(image_content)
        image = Image.open(image_file).convert("RGB")
        plt.imshow(image)
        file_path = Path("C:/Users/akash/OneDrive/Desktop/flipkart dataser/Shoes", "Shoes" +str(i)+ ".png")
        i+=1
        image.save(file_path, "PNG", quality=80)
```
