```python
#This code is used to add the webscraper functionality for fetching trending images on pinterest.

from pinscrape import pinscrape
details = pinscrape.scraper.scrape("lehenga Choli sets for women", "output", {}, 10, 1000)
lst = []

if details["isDownloaded"]:
    print("\nDownloading completed !!")
    print(f"\nTotal urls found: {len(details['extracted_urls'])}")
    print(f"\nTotal images downloaded (including duplicate images): {len(details['url_list'])}")
#     print(details)
    for link in details.get('url_list'):
        lst.append(link)
else:
    print("\nNothing to download !!")
```

```python
lst
```

```python
import requests
import io
from PIL import Image
from pathlib import Path
# hashlib allows you to get hashes. Let's use 'sha1' to name the images.
import hashlib
import matplotlib.pyplot as plt
```

```python
#Below code converts image from ul format to jpg format in RGB form
for c in lst:
        image_content = requests.get(c).content
        image_file = io.BytesIO(image_content)
        image = Image.open(image_file).convert("RGB")
        plt.imshow(image)
        file_path = Path("C:/Users/akash/Documents/pinimages", hashlib.sha1(image_content).hexdigest()[:10] + ".png")
        image.save(file_path, "PNG", quality=80)
```
